```python
import pandas
import numpy as np
from scipy.optimize import curve_fit
from matplotlib.pyplot import plot,xlabel,ylabel,title,legend

df=pandas.read_csv('HemeBinding3-10A3-minus-topnoise.csv')#which file you are using, you need to mess with the CSV some to make it readable.
x1=df.Wavelength10 #which wavelength column you are using.  
y1=df.Abs10 #which absorbance column you are using

x1=x1.values.tolist()#numpy likes things in lists
y1=y1.values.tolist()
y1=np.array(y1)
x1=np.array(x1)
```

```python
def expon(x,a,b,c,d):
    return a*np.exp(-c**x)+b*np.exp(-d*x)
guess=[0.2,8.21048776e+13,0,0.13]
popt, pcov=curve_fit(expon,x1,y1,guess)
print(popt)
```

```python
plot(x1,expon(x1,*popt), label='fitted')
```

```python
plot(x1,y1,label='raw')

```

```python
y2=y1-expon(x1,*popt)
plot(x1,y2,label='raw-fitted')
plot(x1,expon(x1,*popt), label='fitted')
legend (loc="upper left")
```
